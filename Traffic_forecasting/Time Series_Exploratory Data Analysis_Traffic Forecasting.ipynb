#### Moving Average

In this technique, we will take average of passengers counts for last few time periods only instead of taking all the known peiviously values. 
let's try the rolling mean for last 10,20,50 days and visualize the results.

```python
y_hat_avg = cv_data.copy()
# Average of last 10 observtions
y_hat_avg['movingAveForecastWith10'] = train['Count'].rolling(10).mean().iloc[-1]
print(y_hat_avg['movingAveForecastWith10'].head(1))
# Plot it
plt.figure(figsize=(15,8))
plt.plot(train['Count'], label = 'Train')
plt.plot(cv_data['Count'], label = 'Cross Validation')
plt.plot(y_hat_avg['movingAveForecastWith10'], label = 'Moving Average forecasting with 10 observations.')
plt.legend(loc = 'best')
plt.ylabel('#Passangers')
plt.xlabel('Time in Years and Month')
plt.title('Moving Average forecasting with 10 observations')
plt.show()
```

```python
# Average of last 20 observtions
y_hat_avg['movingAveForecastWith20'] = train['Count'].rolling(20).mean().iloc[-1]
print(y_hat_avg['movingAveForecastWith20'].head(1))
# Plot it
plt.figure(figsize=(15,8))
plt.plot(train['Count'], label = 'Train')
plt.plot(cv_data['Count'], label = 'Cross Validation')
plt.plot(y_hat_avg['movingAveForecastWith20'], label = 'Moving Average forecasting with 20 observations.')
plt.legend(loc = 'best')
plt.ylabel('#Passangers')
plt.xlabel('Time in Years and Month')
plt.title('Moving Average forecasting with 20 observations')
plt.show()
```

```python
# Average of last 20 observtions
y_hat_avg['movingAveForecastWith50'] = train['Count'].rolling(50).mean().iloc[-1]
print(y_hat_avg['movingAveForecastWith50'].head(1))
# Plot it
plt.figure(figsize=(15,8))
plt.plot(train['Count'], label = 'Train')
plt.plot(cv_data['Count'], label = 'Cross Validation')
plt.plot(y_hat_avg['movingAveForecastWith50'], label = 'Moving Average forecasting with 50 observations.')
plt.legend(loc = 'best')
plt.ylabel('#Passangers')
plt.xlabel('Time in Years and Month')
plt.title('Moving Average forecasting with 50 observations')
plt.show()
```

Observation : we can see here the predictions are getting weaker with increasing the #datapoints.

```python
rms_moving_avg = sqrt(mean_squared_error(cv_data.Count, y_hat_avg.movingAveForecastWith50))
rms_moving_avg
```

#### Simple Exponential Smoothing
<br></br>
In this technique, we assign larger weight to resecnt observations than to the observations from the distant past.
<br></br>
<br>Note :  If we give the entrire weight to last observed value only then this method will be simila to naive Approach. </br>

<br></br>
<br>Here the prediction is made by assigning larger weight to the recent value and lesser wights to the old values.</br>

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
y_hat_ex = cv_data.copy()
# Model creation
model_SES = SimpleExpSmoothing(np.asarray(train['Count'])).fit(smoothing_level=0.6, optimized= False)
# Forcasting
y_hat_ex['SES'] = model_SES.forecast(len(cv_data))
print(y_hat_ex['SES'].head(1))
```

```python
# Plot it
plt.figure(figsize=(15,8))
plt.plot(train['Count'], label = 'Train')
plt.plot(cv_data['Count'], label = 'Cross Validation')
plt.plot(y_hat_ex['SES'], label = 'SES')
plt.legend(loc = 'best')
plt.ylabel('#Passangers')
plt.xlabel('Time in Years and Month')
plt.title('Simple Exponential Smoothing')
plt.show()
```

Observation : Fit has improved as error has reduced till 113.43708111884514.

```python
rms_SES = sqrt(mean_squared_error(cv_data.Count, y_hat_ex.SES))
rms_SES
```

#### Holt's Linear Trand Model
<br></br>
This technique is extension of Simple exponential smooting to allow forecasting of the data with a trand. 
<br></br>
<br>We acn decompose the Time Series into the four parts,</br>
<br>1. Observed, which is original Time Serires.</br>
<br>2. Trand, which show the trand in Time Series.</br>
<br>3. Seasonal, which tells bout the seasionality in the Time Series.</br>
<br>4. Residual, which is obtained by removing any trand or seasionality in the Time Series.</br>
<br>Lets see it.</br>

```python
import statsmodels.api as sm
```

```python
sm.tsa.seasonal_decompose(train.Count).plot()
results = sm.tsa.stattools.adfuller(train.Count)
plt.show()
```

Increasing Trend is present in data hence we will make model based on trend. 

```python
y_hat_HoltLinear = cv_data.copy()
# Model creation
model_HoltLinear = Holt(np.asarray(train['Count'])).fit(smoothing_level=0.3, smoothing_trend=0.1)
# Forcasting
y_hat_HoltLinear['HoltLinear'] = model_HoltLinear.forecast(len(cv_data))
print(y_hat_HoltLinear['HoltLinear'].head(10))
```

```python
# Plot it
plt.figure(figsize=(15,8))
plt.plot(train['Count'], label = 'Train')
plt.plot(cv_data['Count'], label = 'Cross Validation')
plt.plot(y_hat_HoltLinear['HoltLinear'], label = 'HoltLinear')
plt.legend(loc = 'best')
plt.ylabel('#Passangers')
plt.xlabel('Time in Years and Month')
plt.title('Holt Linear')
plt.show()
```

```python
rms_HoltLinear = sqrt(mean_squared_error(cv_data.Count, y_hat_HoltLinear.HoltLinear))
rms_HoltLinear
```

Observation : Fit has improved as error has reduced from 113.43708111884514 till 112.94278345314041.
